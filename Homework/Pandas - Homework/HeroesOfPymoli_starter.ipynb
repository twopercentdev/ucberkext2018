```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Raw data file
file_to_load = "Resources/purchase_data.csv"

# Read purchasing file and store into pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Establish Variables
uniqueItems = purchase_data['Item Name'].count()
meanPrice = purchase_data['Price'].mean()
purchaseCount = purchase_data['Purchase ID'].count()
revenueTotal = purchase_data['Price'].sum()
#Create Dictionary
sumDict = {'Unique Items' : [uniqueItems] , 'Average Price' : [meanPrice], 'Transaction Count' : [purchaseCount], 'Total Revenue' : [revenueTotal]}
#Create Df
dfSum = pd.DataFrame(data=sumDict).round(2)

dfSum
```

## Gender Demographics

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Establish Variables
genderDemo = purchase_data['Gender'].value_counts()
genderCount = purchase_data['Gender'].count()
#Create Df
dfGender = genderDemo.rename_axis('Gender').reset_index(name='Value Count')
#Create Percentage Row within Df
dfGender['Percentage'] = dfGender.apply(lambda row: (row['Value Count'] / genderCount)*100 , axis=1).round(2)

dfGender
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, etc. by gender


* For normalized purchasing, divide total purchase value by purchase count, by gender


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
dfGenderSum = pd.DataFrame(purchase_data.groupby('Gender')['Price'].sum())
dfGenderTotal = pd.DataFrame(purchase_data.groupby('Gender')['Price'].count())
dfGenderMean = pd.DataFrame(purchase_data.groupby('Gender')['Price'].mean()).round(2)

dfGenderJoin = dfGenderSum.join(dfGenderTotal, lsuffix=' Sum', rsuffix=' Count')
dfGenderFinal = dfGenderJoin.join(dfGenderMean, lsuffix='', rsuffix='')

dfGenderFinal['Normalized Totals'] = dfGenderFinal.apply(lambda row: (row['Price Sum'] / row['Price Count']), axis=1).round(2)
dfGenderFinal.rename(columns={'Price Sum': 'Total Purchase Value', 'Price Count': 'Purchase Count', 'Price': 'Average Purchase Price'})
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data['Age Group'] = pd.cut(purchase_data.Age, age_bins, labels=group_names)

age_demographics = pd.DataFrame(purchase_data['Age Group'].value_counts())
purchase_count = purchase_data['SN'].count()
age_demographics['Percentage'] = age_demographics.apply(lambda row: (row['Age Group'] / purchase_count)*100 , axis=1).round(2)

age_demographics
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, etc. in the table below


* Calculate Normalized Purchasing


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
dfAgeSum = pd.DataFrame(purchase_data.groupby('Age Group')['Price'].sum())
dfAgeTotal = pd.DataFrame(purchase_data.groupby('Age Group')['Price'].count())
dfAgeMean = pd.DataFrame(purchase_data.groupby('Age Group')['Price'].mean()).round(2)

dfAgeJoin = dfAgeSum.join(dfAgeTotal, lsuffix=' Sum', rsuffix=' Count')
dfAgeFinal = dfAgeJoin.join(dfAgeMean, lsuffix='', rsuffix='')

dfAgeFinal['Normalized Totals'] = dfAgeFinal.apply(lambda row: (row['Price Sum'] / row['Price Count']), axis=1).round(2)
dfAgeFinal.rename(columns={'Price Sum': 'Total Purchase Value', 'Price Count': 'Purchase Count', 'Price': 'Average Purchase Price'})
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
dfSnSum = pd.DataFrame(purchase_data.groupby('SN')['Price'].sum())
dfSnTotal = pd.DataFrame(purchase_data.groupby('SN')['Price'].count())
dfSnMean = pd.DataFrame(purchase_data.groupby('SN')['Price'].mean()).round(2)

dfSnJoin = dfSnSum.join(dfSnTotal, lsuffix=' Sum', rsuffix=' Count')
dfSnFinal = dfSnJoin.join(dfSnMean, lsuffix='', rsuffix='')


dfSnFinalSorted = dfSnFinal[['Price Count' , 'Price', 'Price Sum']].sort_values('Price Sum', ascending=False).rename(columns={'Price Sum': 'Total Purchase Value', 'Price Count': 'Purchase Count', 'Price': 'Average Purchase Price'})


dfSnFinalSorted.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
itemdf = purchase_data[['Item Name', 'Item ID', 'Price']]

itemSumDf = pd.DataFrame(itemdf.groupby(['Item ID', 'Item Name'])['Price'].sum())
itemMeanDf = pd.DataFrame(itemdf.groupby(['Item ID', 'Item Name'])['Price'].mean())
itemCountDf = pd.DataFrame(itemdf.groupby(['Item ID', 'Item Name'])['Price'].count())

itemJoin = itemSumDf.join(itemCountDf, lsuffix=' Sum', rsuffix=' Count')
itemJoinFinal = itemJoin.join(itemMeanDf, lsuffix='', rsuffix='')

itemJoinFinalSorted = itemJoinFinal[['Price Count' , 'Price', 'Price Sum']].sort_values('Price Count', ascending=False).rename(columns={'Price Sum': 'Total Purchase Value', 'Price Count': 'Purchase Count', 'Price': 'Average Purchase Price'})
itemJoinFinalSorted.head()
```
